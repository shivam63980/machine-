```python
x=df.drop(columns='diagnosis')
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ssxtrain=ss.fit_transform(xtrain)
ssxtest=ss.fit_transform(xtest)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

```python
yprd=dt.predict(xtest)
```

```python
sc=dt.score(xtest,ytest)
sc
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,yprd)
```

```python
dt.fit(ssxtrain,ytrain)
yprd=dt.predict(ssxtest)
sc=dt.score(ssxtest,ytest)
sc
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
```

```python
yprd=lr.predict(xtest)
sc=lr.score(xtest,ytest)
sc
```

```python
from sklearn.ensemble import RandomForestClassifier
po=RandomForestClassifier()
po.fit(xtrain,ytrain)
```

```python
yprd=po.predict(xtest)
sc=po.score(xtest,ytest)
sc
```
